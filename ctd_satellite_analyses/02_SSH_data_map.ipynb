```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

# Compute June monthly mean for each year
ssh_data = {
    1993: ssh_1993.sla.sel(time=ssh_1993.sla['time.month']==6).mean(dim='time'),
    1998: ssh_1998.sla.sel(time=ssh_1998.sla['time.month']==6).mean(dim='time'),
    2023: ssh_2023.sla.sel(time=ssh_2023.sla['time.month']==6).mean(dim='time')
}

# Create a figure with three panels
fig, axes = plt.subplots(1, 3, figsize=(10, 5),
                         subplot_kw={'projection': ccrs.NorthPolarStereo(central_longitude=-45)})

years = [1993, 1998, 2023]

for ax, year in zip(axes, years):
    data = ssh_data[year]
    
    # Plot the data
    im = data.plot(ax=ax,
                   transform=ccrs.PlateCarree(),
                   vmin=-0.15, vmax=0.15,
                   cmap='RdBu_r',
                   add_colorbar=False)
    
    # Set extent around Greenland
    ax.set_extent([-58, -42, 52, 62], crs=ccrs.PlateCarree())
    
    # Add features
    ax.coastlines(resolution='10m')
    ax.gridlines(draw_labels=False)
    
    ax.set_title(f"SLA June {year}")

# Add a single colorbar below the three panels
cbar = fig.colorbar(im, ax=axes, orientation='horizontal', fraction=0.05, pad=-0.3)
cbar.set_label('SLA (m)')

plt.tight_layout()
plt.show()
```

### Plot trend in SSH mm/yr 1993-2025

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import glob
import dask.array as da
```

```python
# File paths
data_path = "/Users/mzahn/data/SSH/cmems_ssh_yearly/"
out_path = "/Users/mzahn/data/SSH/satellite_ssh_trend_1993_2025_mm_per_year.nc"

files = sorted(glob.glob(data_path + "ssh_cmems_l4_0pt125deg_50N_65N_60W_40W_*.nc"))

# Open lazily with dask
ds = xr.open_mfdataset(
    files,
    combine="by_coords",
    chunks={"time": 365}  # chunk along time
)

# Variable name (adjust if needed)
ssh = ds["sla"]

# Convert time to fractional years
time_years = ds.time.dt.year + (ds.time.dt.dayofyear - 1) / 365.25
ssh = ssh.assign_coords(time=time_years)
```

```python
# Compute linear trend
pf = ssh.polyfit(dim="time", deg=1)

trend = pf["polyfit_coefficients"].sel(degree=1)

# Convert m/year to mm/year
trend_mm_per_year = trend * 1000
trend_mm_per_year.name = "ssh_trend"

trend_mm_per_year.attrs["units"] = "mm yr-1"
trend_mm_per_year.attrs["description"] = "Altimetry-derived mean SSH trend (1993–2025)"
```

```python
# Save NetCDF
trend_mm_per_year.to_netcdf(out_path)
print("Saved trend dataset:", out_path)
```

```python
# Open netcdf
trend_mm_per_year = xr.open_dataset('/Users/mzahn/data/SSH/satellite_ssh_trend_1993_2025_mm_per_year.nc')
```

```python
# Open bathymetry
bathy = xr.open_dataset('/Users/mzahn/data/GEBCO/GEBCO_2023.nc')
bathy_greenland = bathy.elevation.sel(lon=slice(-65,-35),lat=slice(50,70))
```

```python
plt.rcParams['font.size'] = 14

# Plot map
lon = ds["longitude"]
lat = ds["latitude"]

fig = plt.figure(figsize=(8,7))
ax = plt.axes(projection=ccrs.NorthPolarStereo(central_longitude=-45))
ax.set_extent([-60, -40, 52, 65], crs=ccrs.PlateCarree())  # Greenland region

# Contour-filled SSH trend
# levels = np.arange(-6, 6, 1)
levels = np.arange(1, 5, 0.1)

cf = ax.contourf(
    lon,
    lat,
    trend_mm_per_year.ssh_trend,
    levels=levels,
    cmap="RdBu",
    extend="both",
    transform=ccrs.PlateCarree()
)

# Colorbar
cbar = plt.colorbar(cf, ax=ax, orientation="vertical", pad=0.05)
cbar.set_label("SSH trend (mm/yr)")

# Coastlines and land
ax.coastlines()
ax.add_feature(cfeature.LAND, color="lightgray")

# Bathymetry contours
bathy_levels = [-3000, -1000]

cs = ax.contour(
    bathy_greenland.lon,
    bathy_greenland.lat,
    bathy_greenland,
    levels=bathy_levels,
    colors="black",
    linewidths=0.8,
    linestyles="solid",
    transform=ccrs.PlateCarree()
)

# ax.clabel(cs, fmt=lambda x: f"{abs(int(x))} m", fontsize=8)

plt.title("Mean SLA Trend (1993–2025)")

plt.show()
```
